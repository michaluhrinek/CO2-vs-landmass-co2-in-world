CO2 per Landmass

Import libraries 

```python
import pandas as pd 
```

Load dataset 

```python
land_df = pd.read_csv('data.csv')
```

```python
land_df
```

drop column place

```python
land_df.drop(columns=['place'], axis=1, inplace=True)
```

```python
land_df
```

Drop columns from dataset which are not needed

```python
land_df.drop(columns=['pop2023', 'growthRate', 'area','cca3','cca2','ccn3','densityMi','Rank','rank'], axis=1, inplace=True)
```

```python
land_df
```

```python
co2_df = pd.read_csv("https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv")
```

```python
co2_df
```

```python
co2_df.dropna(inplace=True)
```

```python
co2_df
```

Clean columns

```python
co2_df.drop(columns=['co2_growth_abs','co2_growth_prct','share_global_cumulative_other_co2','share_global_flaring_co2','share_global_gas_co2','trade_co2_share','trade_co2','total_ghg_excluding_lucf','total_ghg','share_global_other_co2','share_global_oil_co2','share_global_luc_co2'], axis=1, inplace=True)
```

```python
co2_df
```

Merge 2data sets together 

```python
merged_df = pd.merge(co2_df, land_df, on="country")
```

```python
merged_df
```

Clean NAN values 

```python
merged_df.dropna(inplace=True)
```

```python
merged_df
```

Calculate emissions per unit/ per landmass in km

```python
merged_df['co2_per_landmass'] = merged_df['co2'] / merged_df['landAreaKm']  #co2 on km2
```

```python
merged_df
```

save data, new data into new dataframe 

```python
merged_df.to_csv("dataco2.csv", index=False)
```
